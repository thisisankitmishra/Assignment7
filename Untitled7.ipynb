Question 1: What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
Answer:
Artificial Intelligence (AI) is the field of building systems that can mimic human intelligence, reasoning, and decision making. Applications include chatbots, robotics, and autonomous vehicles.

Machine Learning (ML) is a subset of AI that focuses on algorithms that learn from data to make predictions. Techniques include regression, classification, clustering. Examples: spam detection, recommendations.

Deep Learning (DL) is a subset of ML based on multi-layer neural networks that learn complex patterns. Techniques include CNN, RNN. Applications include image recognition, speech processing.

Data Science focuses on collecting, cleaning, analyzing, visualizing data, and building predictive models to support decision-making. It uses statistics, ML, SQL, visualization. Applications include forecasting, business analytics.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
Answer:
Overfitting occurs when a model learns noise and performs well on training data but poorly on test data. Underfitting occurs when a model is too simple and fails to learn useful patterns.

Detection: Compare training and test accuracy, cross-validation, learning curves.

Prevention:
- Cross-validation
- Regularization (L1, L2)
- Reduce model complexity
- Early stopping
- Dropout (in deep learning)

Overfitting = high variance, Underfitting = high bias. Aim to balance both.

Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples.
Answer:
1. Deletion:
Remove rows/columns with missing values when missing percentage is small.
Example: df.dropna()

2. Mean/Median Imputation:
Replace missing numeric values with mean or median to preserve dataset size.
Example: df["Age"].fillna(df["Age"].median())

3. Predictive Modeling:
Use machine learning models to predict missing values using other variables.
Example: regression-based imputation for missing salary.

Question 4: What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
Answer:
An imbalanced dataset occurs when one class has far more samples than others (e.g., 95% normal, 5% fraud).

Techniques:
1. SMOTE (Oversampling):
Creates synthetic minority samples to balance data.

2. Random Undersampling:
Reduces majority class samples to match minority class.

Other techniques include class weighting and balanced metrics.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.
Answer:
Feature scaling is important because many algorithms use distance or gradient-based learning, and unscaled features can dominate learning.

Min-Max Scaling:
Transforms features to range 0–1. Sensitive to outliers. Useful in neural networks.

Standardization:
Transforms features to mean 0 and standard deviation 1. Less affected by outliers. Useful for SVM, KNN, logistic regression.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Answer:
Label Encoding maps categories to numeric values. Suitable for ordinal data (e.g., low, medium, high).

One-Hot Encoding creates binary columns for each category. Suitable for nominal data (e.g., city, color).

Prefer Label Encoding for ordered categories.
Prefer One-Hot Encoding when categories have no order.

Question 7: Google Play Store Dataset

Python Code:
import pandas as pd
df = pd.read_csv("googleplaystore.csv")
result = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)
print(result)

Answer:
Categories like Education, Health & Fitness, and Productivity have the highest average ratings. Categories like Entertainment, Dating, and Tools have the lowest ratings.

Reasons:
High ratings due to usefulness and reliability.
Low ratings due to bugs, ads, and subjective satisfaction.
